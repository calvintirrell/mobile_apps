# Import required packages to work with and visualize data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

# Load and preview the 2 data files

```python
app_data = pd.read_csv('AppleStore.csv')
display(app_data)
```

```python
descriptions_data = pd.read_csv('appleStore_description.csv')
display(descriptions_data)
```

# Combine the 2 separate data sets into 1 data set

```python
merged_app_data = pd.merge(app_data, descriptions_data, how = 'left', left_on = ['id','track_name'], right_on = ['id','track_name'])
merged_app_data.tail()
```

# Drop the un-needed columns from the combined data set

```python
merged_app_data.drop(labels=['Unnamed: 0', 'size_bytes_y'], axis=1, inplace=True)
```

```python
merged_app_data.head()
```

# Rename the column names to be easier to understand

```python
# Current column names to be renamed; so will stay the same though
merged_app_data.columns
```

```python
merged_app_data.rename(columns = {'id':'app_id_num', 'track_name':'app_name', 'size_bytes_x':'app_size_bytes', 'currency':'currency', 'price':'price',
                       'rating_count_tot':'total_rating_count', 'rating_count_ver':'version_rating_count', 'user_rating':'user_rating',
                       'user_rating_ver':'user_rating_version', 'ver':'app_version', 'cont_rating':'content_rating', 'prime_genre':'main_genre',
                       'sup_devices.num':'supported_devices_count', 'ipadSc_urls.num':'screenshots_shown', 'lang.num':'supported_languages',
                       'vpp_lic':'device_licensing', 'app_desc':'app_description'}, inplace = True)
```

```python
# Showing the new column names
display(merged_app_data)
```

```python
app_data_ratings = merged_app_data.groupby('app_name', as_index=False)['total_rating_count'].max().sort_values('total_rating_count', ascending=False)
app_data_ratings.columns = ['App Name','App Ratings']
print('Top 5 apps with the most ratings:')
app_data_ratings.head()
```

```python
print('This doesn\'t really tell the whole story of apps with no ratings: ')
app_data_ratings.tail()
```

```python
print('Apps with 0 ratings:')
app_data_ratings[app_data_ratings['App Ratings']==0].count()
```

```python
print('Apps with 1 rating:')
app_data_ratings[app_data_ratings['App Ratings']==1].count()
```

```python
print('Apps with 7 ratings:')
app_data_ratings[app_data_ratings['App Ratings']==7].count()
```

```python
print('929 or about 12% of all apps have 0 ratings as is shown above as well')
count_ratings = app_data_ratings['App Ratings'].value_counts(ascending=False).head().to_frame()
percent_ratings = app_data_ratings['App Ratings'].value_counts(ascending=False, normalize=True).head().to_frame()
side_by_side = pd.concat([count_ratings, percent_ratings], axis=1)
side_by_side
```

# Correlation heatmap of data columns

```python
correlations = merged_app_data.corr()
f, ax = plt.subplots(figsize=(16, 8))
mask = np.zeros_like(correlations, dtype=np.bool)
mask[np.triu_indices_from(mask)]=True

heatmap = sns.heatmap(correlations, mask=mask, cmap='vlag_r')
print("The strongest correlation between any 2 columns is between 'user_rating' and 'user_rating_version'.")
print("The next strongest correlations seems to a tie between the pairing of 'user_rating' and 'screenshots_shown' ")
print("as well as the pairing of 'user_rating_version' and 'screenshots_shown'.")
print("I ignore the 'app_id_num' column in this case as the ID number of an app shouldn't impact its usage or ratings, etc.")
print(heatmap)
```

# Let's look into App Prices

```python
plt.figure(figsize=(16,8))
plt.xlabel('App Prices Distribution in USD ($)')
plt.ylabel('Frequency of App Price (log scale)')
plt.hist(data=merged_app_data, x='price', log=True, color='green')
```

```python
plt.figure(figsize=(16,8))
plt.title('Price Distribution Visual')
sns.stripplot(data=merged_app_data, x='price', jitter= True, orient='h', size=3)
plt.show()
```

# Create 2 groups: paid and free apps

Let's look at free apps first!

```python
free_apps = merged_app_data[merged_app_data['price']==0]
print("Number of free apps:")
free_apps.count()
```

```python
f, ax = plt.subplots(figsize=(16, 8))
plt.title('Free Apps User Rating Distribution')
sns.countplot(x="user_rating", data=free_apps, palette="Reds")
```

```python
print('From the free apps, look into the free games:')
only_free_games = free_apps[free_apps['main_genre']=='Games']
only_free_games.head(3)
```

```python
print('From the free apps, look into the free games:')
only_free_games.describe()
```

```python
print('Average user rating for free games is: 3.53. Median of user ratings for free games is: 4.0.')

f, ax = plt.subplots(figsize=(10, 10))
plt.title('Free Games User Rating Distribution')
sns.boxplot(data=only_free_games, x='main_genre', y='user_rating')
```

Now let's look into the paid apps!

```python
paid_apps = merged_app_data[merged_app_data['price']>0]
print("Number of paid apps:")
paid_apps.count()
```

```python
f, ax = plt.subplots(figsize=(16, 8))
plt.title('Paid Apps User Rating Distribution')
sns.countplot(x="user_rating", data=paid_apps, palette="Blues")
```

```python
print('From the paid apps, look into the paid games:')
only_paid_games = paid_apps[paid_apps['main_genre']=='Games']
only_paid_games.head(3)
```

```python
print('From the paid apps, look into the paid games:')
only_paid_games.describe()
```

```python
print('Average user rating for paid games is: 3.9. Median and 75th percentile of user ratings for paid games overalp at: 4.5.')

f, ax = plt.subplots(figsize=(10, 10))
plt.title('Paid Games User Rating Distribution')
sns.boxplot(data=only_paid_games, x='main_genre', y='user_rating', color='green')
```
